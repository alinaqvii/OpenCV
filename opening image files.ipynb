# Opening Image file in Notebook

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
img = cv2.imread('/home/ali/Desktop/COURSES/skardu pics 025.JPG')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# MATPLOTLIB --> RGB
# OPEN CV -----> BGR
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_grey = cv2.imread('/home/ali/Desktop/COURSES/skardu pics 025.JPG', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_grey, cmap='gray')
```

```python
fix_img.shape
```

```python
new_size = cv2.resize(fix_img, (1000,400))
```

```python
plt.imshow(new_size)
```

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, 1)
plt.imshow(new_img)
```

```python
cv2.imwrite('Totally_new_img.jpg', fix_img)
```

```python
fig = plt.figure(figsize=(10,8)) #to display and adjust canvas size of screen
ax = fig.add_subplot(111)
ax.imshow(fix_img)
```
